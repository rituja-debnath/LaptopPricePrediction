```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('LaptopPricePrediction.csv')
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
#checking if any null value is present or not
df.isnull().sum()
```

```python
#changing the format of how price is given
df['Price'] = df['Price'].str.replace(",", "")
df['Price'] = df['Price'].str.replace("₹", "")
```

```python
df.head()
```

```python
#representation of all the different types of RAM 
sns.countplot(y = df['RAM'])
```

```python
#Representation of the different types of processors present in the dataset 
sns.countplot(y= df['Processor'])
```

```python
#Representation of different displays
sns.countplot(y = df['Display'])
```

```python
#changing format of RAM in dataset by keeping just the numeric part
df['RAM'] = df['RAM'].str.replace('\s.*', '')
```

```python
df
```

```python
#changing the format of representation of warranty in dataset by keeping just the numeric part
df['Warranty'] = df['Warranty'].str.replace('\s.*', '')
```

```python
df.head()
```

```python
#rating is not affecting the price column and it has been observed that almost all of them has same operating system so dropping operating system
df.drop(['rating', 'Operating System'], axis = 1, inplace = True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lr=LabelEncoder()
```

```python
#encoding numeric values to all of them
df['Name']=lr.fit_transform(df['Name'])
df['Processor']=lr.fit_transform(df['Processor'])
df['Storage']=lr.fit_transform(df['Storage'])
df['Display']=lr.fit_transform(df['Display'])
df['RAM']=lr.fit_transform(df['RAM'])
df['Warranty']=lr.fit_transform(df['Warranty'])
df['Price']=df['Price'].astype(str).astype(int)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
x = df.drop(['Price'], axis = 1)
y = df['Price'] #seperating the target one
```

```python
x
```

```python
y
```

```python
#splitting data into test and train 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
```

```python
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(x_train,y_train)
```

```python
training_data_prediction = lin_reg_model.predict(x_train)
```

```python
error_score = metrics.r2_score(y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
training_data_prediction=lin_reg_model.predict(x_test)
error_score=metrics.r2_score(y_test,training_data_prediction)
print("R Squared Error : ",error_score)
```

```python
las_reg_model=Lasso()
las_reg_model.fit(x_train,y_train)
```

```python
training_data_prediction=las_reg_model.predict(x_train)
error_score=metrics.r2_score(y_train,training_data_prediction)
print("R Squared Error : ",error_score)
```

```python
training_data_prediction=las_reg_model.predict(x_test)
error_score=metrics.r2_score(y_test,training_data_prediction)
print("R Squared Error : ",error_score)
```
